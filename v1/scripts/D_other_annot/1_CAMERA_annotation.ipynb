```r
ms1_data <- filterMsLevel(data$spectra_centroid, msLevel = 1L)
xs <- as(ms1_data, 'xcmsSet')
```

```r
sampclass(xs) <- pData(data$spectra_centroid)$treatment
sampnames(xs) <- pData(data$spectra_centroid)$SampleID
```

```r
an <- xsAnnotate(xs)
anF <- groupFWHM(an, perfwhm = 0.6)
anI <- findIsotopes(anF, mzabs = 0.01)
anIC <- groupCorr(anI, cor_eic_th = 0.75)
anFA <- findAdducts(anIC, polarity="positive")
```

```r
peaklist <- getPeaklist(anFA)
write_csv(peaklist, file=file.path(tables_dir, 'xsannotated.csv'))
```
